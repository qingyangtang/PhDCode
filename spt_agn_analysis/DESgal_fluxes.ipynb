```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
sumss5 = np.asarray([16.2, 21.5, 9.8, 11.9, 16.5, 7.7])
sumss23 = np.asarray([10.0,18.9,11.7,8.4,14.5,8.6])
DES5 = np.asarray([1.7, 9.8, 2.8, 2.5, 7.7, 2.1])
DES23 = np.asarray([-3.7,6,5.1,-3.6,2.3,1.5])
DES23z1 = np.asarray([-1.8,2.9,2.6,-1.5,1.6,0.3])
DES23z2 = np.asarray([-2.6,5.9,4.8,-2.8,1.9,2.5])
DES23z3 = np.asarray([-2.1,0.2,0.3,-2.2,-0.1,-1.3])
DES5z1 = np.asarray([2.1,5.1,-0.1,2.9,4.0,-0.2])
DES5z2 = np.asarray([0.4,7.5,3.6,1.1,7.0,3.0])
DES5z3 = np.asarray([0.7,3.4,0.1,0.9,1.0,-0.3])
```

```python
freq = np.asarray([90,150,220])
color = ['black','red','blue','purple','green']
plt.figure(figsize=(20,9))
plt.plot(freq, sumss23[:3], color = color[0], label='SUMSS')
plt.plot(freq, DES23[:3], color = color[1],label='DES, all z')
plt.plot(freq, DES23z1[:3],color = color[2],label='DES, z<0.25')
plt.plot(freq, DES23z2[:3],color = color[3],label='DES 0.25<z<0.75')
plt.plot(freq, DES23z3[:3],color = color[4],label='DES z>0.75')
plt.plot(freq, sumss23[3:],'--',color = color[0], label='SUMS med')
plt.plot(freq, DES23[3:],'--',color = color[1], label='DES, all z, med')
plt.plot(freq, DES23z1[3:],'--',color = color[2],label='DES, z<0.25 med')
plt.plot(freq, DES23z2[3:],'--',color = color[3],label='DES 0.25<z<0.75 med')
plt.plot(freq, DES23z3[3:],'--',color = color[4],label='DES z>0.75 med')
plt.legend(loc='best')
plt.xlabel("Freq (GHz)")
plt.ylabel("S/N")
plt.title("S/N for 23h map")
plt.show()
```
